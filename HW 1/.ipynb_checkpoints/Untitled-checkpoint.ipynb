```python
import os
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
from math import sqrt
```

```python
def contrast_scratching(pic, func):
    new_image = np.zeros(pic.shape, pic.dtype)
    
    if func == "first":
        for x in range(pic.shape[0]):
            for y in range(pic.shape[1]):
                for z in range(pic.shape[2]):
                    new_image[x,y,z] = np.clip(pic[x,y,z] * -1 + 1, 0, 1)
        return new_image * 255
    
    elif func == "second":
        for x in range(pic.shape[0]):
            for y in range(pic.shape[1]):
                for z in range(pic.shape[2]):
                    if pic[x,y,z] >= 0.2 and pic[x,y,z] <= 0.55:
                        new_image[x,y,z] = np.clip(0, 0, 1)
                    else:
                        new_image[x,y,z] = np.clip(pic[x,y,z], 0, 1)
        return new_image * 255
    
    elif func == "third":
        for x in range(pic.shape[0]):
            for y in range(pic.shape[1]):
                for z in range(pic.shape[2]):
                    if pic[x,y,z] >= 0.4 and pic[x,y,z] <= 0.55:
                        new_image[x,y,z] = np.clip(pic[x,y,z], 0, 1)
                    elif pic[x,y,z] > 0.55:
                        new_image[x,y,z] = np.clip(1, 0, 1)
                    elif pic[x,y,z] < 0.4:
                        new_image[x,y,z] = np.clip(0, 0, 1)
        return new_image * 255
    
    elif func == "forth":
        for x in range(pic.shape[0]):
            for y in range(pic.shape[1]):
                for z in range(pic.shape[2]):
                    if pic[x,y,z] >= 0.2 and pic[x,y,z] <= 0.55:
                        new_image[x,y,z] = np.clip(0.3, 0, 1)
                    elif pic[x,y,z] > 0.55:
                        new_image[x,y,z] = np.clip(pic[x,y,z] * 8/9 + 1/9, 0, 1)
                    elif pic[x,y,z] < 0.2:
                        new_image[x,y,z] = np.clip(pic[x,y,z] * 1.5, 0, 1)
        return new_image * 255
    
    elif func == "2thRoot":
        for x in range(pic.shape[0]):
            for y in range(pic.shape[1]):
                for z in range(pic.shape[2]):
                    new_image[x,y,z] = np.clip(sqrt(pic[x,y,z]), 0, 1)
        return new_image * 255
    
    elif func == "2thPower":
        for x in range(pic.shape[0]):
            for y in range(pic.shape[1]):
                for z in range(pic.shape[2]):
                    new_image[x,y,z] = np.clip(pic[x,y,z] ** 2, 0, 1)
        return new_image * 255
```

```python
pic = cv.imread("./assets/img1.tif")
functions = ["first", "second", "third", "forth", "2thRoot", "2thPower"]

path = "C:/Users/98912/Desktop/uni/multimedia/HW1/results/q1"
stretched_pics = []

for i in functions:
    new_pic = pic.copy()/255
    new_pic = contrast_scratching(new_pic, i).astype(int)
    new_pic = new_pic.astype(int)
    stretched_pics.append(new_pic)
    
    name = "{name}.jpeg".format(name = i)
    cv.imwrite(os.path.join(path, name), new_pic)

fig, axs = plt.subplots(1,6,figsize=(16,9))
for i in range(len(functions)):
    axs[i].imshow(stretched_pics[i])
    axs[i].set_title(functions[i])
    
```

```python
# load image
img = cv.imread('./assets/img3.jpg')

# convert to gray
gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

# threshold input image
thresh = cv.threshold(gray, 0, 255, cv.THRESH_BINARY+cv.THRESH_OTSU)[1]

# apply morphology open and close and dilate
kernel = cv.getStructuringElement(cv.MORPH_ELLIPSE, (11,11))
mask = cv.morphologyEx(thresh, cv.MORPH_OPEN, kernel)
# kernel = cv.getStructuringElement(cv.MORPH_ELLIPSE, (55,55))
# mask = cv.morphologyEx(mask, cv.MORPH_CLOSE, kernel)
kernel = cv.getStructuringElement(cv.MORPH_ELLIPSE, (3,3))
mask = cv.morphologyEx(mask, cv.MORPH_DILATE, kernel)

copy_mask = mask
# invert mask
mask = 255 -mask

# make mask 3 channel
mask = cv.merge([mask,mask,mask])

# apply mask to image to blacken circle area
result = img.copy()
result = cv.bitwise_and(img, mask)
mask_inv = cv.bitwise_not(mask)


# # save resulting masked image
cv.imwrite('./results/q3/white_circle_thresh.jpg', thresh)
cv.imwrite('./results/q3/white_circle_mask.jpg', mask_inv)
cv.imwrite('./results/q3/white_circle_masked.jpg', result)

# display original image
fig, ax = plt.subplots(1,1,figsize=(8,5))
ax.set_title("Original picture")
ax.imshow(img)
cv.waitKey(0)
cv.destroyAllWindows()

# display result, though it won't show transparency
fig, ax = plt.subplots(1, 3, figsize=(15,10))
ax[0].imshow(thresh)
ax[0].set_title("Threshold")
ax[1].imshow(mask_inv)
ax[1].set_title("Circles")
ax[2].imshow(result)
ax[2].set_title("lines")
cv.waitKey(0)
cv.destroyAllWindows()
```

```python
import os
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
# load image
# img = cv.imread('./assets/img4.jpg', cv.IMREAD_GRAYSCALE)
img = cv.imread('./assets/img4.jpg')

img = np.asanyarray(img)

# convert to gray
img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)

# Remove noise by blurring with a Gaussian filter
gaussian = cv.GaussianBlur(img,(7,7),0)
median = cv.medianBlur(img,7)
bilateral = cv.bilateralFilter(img,7,75,75)
# fig, axs = plt.subplots(1,3, figsize=(15,10))
# axs[0].imshow(gaussian)
# axs[0].set_title("Guassian Filter")
# axs[1].imshow(median)
# axs[1].set_title("Median Filter")
# axs[2].imshow(bilateral)
# axs[2].set_title("Bilateral Filter")

# Edge detection
edges_G = cv.Canny(gaussian,200,200)
edges_M = cv.Canny(median,200,200)
edges_B = cv.Canny(bilateral,200,00)

# image showing
fig, ax = plt.subplots(2,2, figsize=(15,10))
ax[0,0].imshow(img,cmap = 'gray')
ax[0,0].set_title("Original Image")

ax[0,1].imshow(edges_G,cmap = 'gray')
ax[0,1].set_title('Edge Image - Gaussian Filter')

ax[1,0].imshow(edges_M,cmap = 'gray')
ax[1,0].set_title("Edge Image - Median Filter")

ax[1,1].imshow(edges_B,cmap = 'gray')
ax[1,1].set_title('Edge Image - Bilateral Filter')

# Saving Edge pictures in results
path = "C:/Users/98912/Desktop/uni/multimedia/HW1/results/q4"
edges_pics = [edges_G, edges_M, edges_B]
filters = ["Gaussian Filter", "Median Filter", "Bilateral Filter"]
for pic, name in zip(edges_pics, filters):    
    name = "Canny {name}.jpeg".format(name = name)
    cv.imwrite(os.path.join(path, name), pic)
```
