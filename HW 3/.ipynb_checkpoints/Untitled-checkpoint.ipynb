## Face detection - image

```python
import cv2
import matplotlib.pyplot as plt
```

```python
# Reading the image
img = cv2.imread('faces.jpg')
  
# Converting image to grayscale
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  
# Loading the required haar-cascade xml classifier file
haar_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "Haarcascade_frontalface_default.xml")
  
# Applying the face detection method on the grayscale image
faces_rect = haar_cascade.detectMultiScale(gray_img, 1.1, 8)
  
# Iterating through rectangles of detected faces
cnt = 0
for (x, y, w, h) in faces_rect:
    cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)
    cnt += 1
    
print("Number of detected faces = {cnt}".format(cnt=cnt))    

cv2.imshow('Detected faces', img)
cv2.waitKey(0)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
ax.imshow(img)
```

```python
# Reading the image
img = cv2.imread('faces.jpg')
  
# Converting image to grayscale
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  
# Loading the required haar-cascade xml classifier file
haar_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "Haarcascade_frontalface_alt2.xml")
  
# Applying the face detection method on the grayscale image
# faces_rect = haar_cascade.detectMultiScale(gray_img, 1.1, 9)
faces_rect = haar_cascade.detectMultiScale(gray_img, scaleFactor=1.03, minNeighbors=8, 
                                           minSize=(20, 20), flags=cv2.CASCADE_SCALE_IMAGE)

  
# Iterating through rectangles of detected faces
cnt = 0
for (x, y, w, h) in faces_rect:
    cv2.rectangle(img, (x, y), (x+w, y+h), (0, 255, 0), 2)
    cnt += 1
    
print("Number of detected faces = {cnt}".format(cnt=cnt))    
 
cv2.imshow('Detected faces', img)
  
cv2.waitKey(0)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(15, 10))
ax.imshow(img)
```

## Face detection - Video

```python
import cv2
import time
import imutils
from imutils.video import VideoStream
from matplotlib import pyplot

detector = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_alt2.xml")

cap= cv2.VideoCapture("IMG_2751.mp4")

cnt = 0
start = time.time()

while True:
    frame = cap.read()
    
    if frame[0] == False:
        break
    
    frame = imutils.resize(frame[1], width=720)
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    rects = detector.detectMultiScale(gray, scaleFactor=1.25, minNeighbors=5,
                                      minSize=(20, 20),flags=cv2.CASCADE_SCALE_IMAGE)

    for (x, y, w, h) in rects:
        cv2.rectangle(frame, (x, y), (x + w, y + h), (0, 0, 255), 2)

    cv2.imshow("Draiz Video", frame)
    
    k = cv2.waitKey(30) & 0xff
    if k == 27: # press 'ESC' to quit
        break
    
    if cnt == 200: # Print average fps for the last 200 frames
        end = time.time()
        print(frame.shape)
        print("FPS: ", 200/(end-start))
        c = 0
        start = time.time()
    cnt += 1

cv2.destroyAllWindows()
```
